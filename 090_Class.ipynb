```python
from dataclasses import dataclass


@dataclass
class Coordinate:
    long:float
    lat:float


    def geo_address(self):
        """Call a web service to fetch location by coordinates"""
        ...


@dataclass(frozen=True,slots=True)
class Run:
    milage: float
    duration: float
    location: Coordinate
    description: str | None = None

    def speed(self) -> float:
        return self.milage / (self.duration/60)

    def pace(self) -> float:
        return self.duration/self.milage
```

```python
r1 = Run(
        4.38,
        37.37,
        location=(40.2, 36.3),
        description="Tuesday Run",
    )

r1.speed()

r1.pace()
```

```python
r2 = Run(
        milage=4.82,
        duration=41.48,
        location=(40.2, 36.3),
        description="Morning run",
    )
```

```python
class RunApp:
    runs: list[Run]

    def __init__(self)-> None:
        self.runs = []

    def add_run(self, r: Run):
        self.runs.append(r)

    def best_run(self):
        return min(self.runs, key=lambda r: r.pace())

    def longest_run(self):
        return max(self.runs, key=lambda r: r.milage)
```

```python
app = RunApp()
app.add_run(r1)
app.add_run(r2)
```

```python
app.best_run()
```

```python
app.longest_run()
```
