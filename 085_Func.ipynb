```python
from math import fabs,sqrt
from typing import Callable

y_true = [1, 2, 3, 4, 5, 6, 7, 8, 0]
y_pred = [1, 2, 4, 5, 6, 9, 8, 3, 0]
```

```python
def _mean(l:list) -> float:
    n = len(l)

    if n == 0:
        raise ValueError("Number of elements can not be 0.")
        
    return sum(l) /n
    
def mae(y_true: list[int], y_pred:list[int], w: list[float]|None=None) -> float:
    return _mean([fabs(a-b) for a,b in zip(y_true,y_pred,strict=True)])

def rmse(y_true: list[int], y_pred:list[int], w: list[float]|None=None) -> float:
    return sqrt(_mean([(a-b)**2 for a,b in zip(y_true,y_pred,strict=True)]))


def jaccard(y_true: list[int], y_pred:list[int], w: list[float]|None=None) -> float:
    return 1- len( set(y_true) & set(y_pred)) / len( set(y_true) | set(y_pred))

    
```

```python
mae(y_true, y_pred)
```

```python
rmse(y_true, y_pred)
```

```python
jaccard(y_true, y_pred)
```

```python
scoring: dict[str, Callable[[list,list], float]] = {"mae": mae, "rmse": rmse,"jacc": jaccard}
```

```python
def score(metric:str, y_true, y_pred)->float:
    fn = scoring.get(metric, mae)

    return fn(y_true,y_pred)
```

```python
score("mae", [1,2],[5,4])
```

```python
def super_metric(metric_names:list[str], weights: list[float], y_true: list[int], y_pred:list[int],) -> float:

    total_weights= sum(weights)
    _weights = [w/total_weights for w in weights]


    return sum( (w * score(name, y_true, y_pred) for w, name in zip(_weights, metric_names,strict=True) ) )
```

```python
super_metric( ["mae","rmse"] ,[2,1], [1,2],[5,4])
```

```python
def super_metric_2(metrics: Callable[[list,list], float], weights: list[float], y_true: list[int], y_pred:list[int],) -> float:

    total_weights= sum(weights)
    _weights = [w/total_weights for w in weights]


    return sum( (w * metric( y_true, y_pred) for w, metric in zip(_weights, metrics,strict=True) ) )
```

```python
super_metric( [mae,rmse] ,[2,1], [1,2],[5,4])
```
