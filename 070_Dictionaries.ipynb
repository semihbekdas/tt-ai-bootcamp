```python
text = "The quick brown fox jumps over the lazy dog"
```

```python
i2s = []
s2i = {}
i=0
for token in text.split():
    if token.lower()  not in i2s:
        i2s.append(token.lower())

        s2i[token.lower()] = i

        i +=1

i2s, s2i
```

```python

text1 = "The quick brown fox jumps over the lazy dog"
text2 = "The quick fox jumps over the lazy dog"
text3 = "The  fox jumps over the  dog"
text4 = "The  fox jumps"

texts = [text1,text2,text3,text4]

texts
```

```python
d = {}
s = {}


for i, doc in enumerate(texts):

    tokens = [token.lower() for token in doc.split()]
    for j, word in enumerate(i2s):
        if word in tokens:
            d[i,j] = len(word)
            s[i,j] = len(word)
        else:
            d[i,j] = 0
        
```
